```python

# División de datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split

# División 70/30 (puedes cambiar a 80/20 si prefieres)
X_train, X_test, y_train, y_test = train_test_split(
    x,  # Variables independientes (ya transformadas)
    y,  # Variable dependiente (ya codificada)
    test_size=0.3,  # 30% para prueba, 70% para entrenamiento
    random_state=42,  # Para reproducibilidad
    stratify=y  # Mantiene la proporción de clases en ambos conjuntos
)

# Verificar las dimensiones
print("Dimensiones de los conjuntos:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")

# Verificar la proporción de clases en cada conjunto
train_proportion = pd.Series(y_train).value_counts(normalize=True).sort_index()
test_proportion = pd.Series(y_test).value_counts(normalize=True).sort_index()

print("\nProporción de clases en entrenamiento:")
print(f"No Churn (0): {train_proportion[0]:.3f}")
print(f"Churn (1): {train_proportion[1]:.3f}")

print("\nProporción de clases en prueba:")
print(f"No Churn (0): {test_proportion[0]:.3f}")
print(f"Churn (1): {test_proportion[1]:.3f}")

print(f"\n✅ División completada exitosamente")
print(f"📊 Total de muestras: {len(x)}")
print(f"🎯 Entrenamiento: {len(X_train)} muestras ({len(X_train)/len(x)*100:.1f}%)")
print(f"🧪 Prueba: {len(X_test)} muestras ({len(X_test)/len(x)*100:.1f}%)")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================================
# MODELO 1: KNN (K-Nearest Neighbors) - REQUIERE NORMALIZACIÓN
# =========================================================================

print("🤖 MODELO 1: K-NEAREST NEIGHBORS (KNN)")
print("="*50)

# Normalizar los datos para KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo KNN
knn_model = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_model.fit(X_train_scaled, y_train)

# Predicciones
y_pred_knn = knn_model.predict(X_test_scaled)
y_pred_proba_knn = knn_model.predict_proba(X_test_scaled)[:, 1]

# Métricas KNN
accuracy_knn = accuracy_score(y_test, y_pred_knn)
roc_auc_knn = roc_auc_score(y_test, y_pred_proba_knn)

print(f"Exactitud (Accuracy): {accuracy_knn:.4f}")
print(f"ROC-AUC Score: {roc_auc_knn:.4f}")
print("\nReporte de clasificación KNN:")
print(classification_report(y_test, y_pred_knn))
```

```python
print("\n🌳 MODELO 2: ÁRBOL DE DECISIÓN")
print("="*50)

# Crear y entrenar el modelo de Árbol de Decisión
tree_model = DecisionTreeClassifier(
    random_state=42,
    max_depth=10,  # Evita overfitting
    min_samples_split=20,  # Mínimo de muestras para dividir
    min_samples_leaf=10,   # Mínimo de muestras por hoja
    class_weight='balanced'  # Balancea las clases automáticamente
)
tree_model.fit(X_train, y_train)  # Sin normalización

# Predicciones
y_pred_tree = tree_model.predict(X_test)
y_pred_proba_tree = tree_model.predict_proba(X_test)[:, 1]

# Métricas Árbol de Decisión
accuracy_tree = accuracy_score(y_test, y_pred_tree)
roc_auc_tree = roc_auc_score(y_test, y_pred_proba_tree)

print(f"Exactitud (Accuracy): {accuracy_tree:.4f}")
print(f"ROC-AUC Score: {roc_auc_tree:.4f}")
print("\nReporte de clasificación Árbol de Decisión:")
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
def evaluar_modelo(y_true, y_pred, y_pred_proba, modelo_nombre):
    """
    Calcula todas las métricas de evaluación para un modelo
    """
    metrics = {
        'Modelo': modelo_nombre,
        'Exactitud': accuracy_score(y_true, y_pred),
        'Precisión': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred),
        'ROC-AUC': roc_auc_score(y_true, y_pred_proba)
    }
    return metrics
```

```python
print("📊 EVALUACIÓN DETALLADA DE MODELOS")
print("="*60)

# Calcular métricas para KNN
metrics_knn = evaluar_modelo(y_test, y_pred_knn, y_pred_proba_knn, 'KNN')

# Calcular métricas para Árbol de Decisión
metrics_tree = evaluar_modelo(y_test, y_pred_tree, y_pred_proba_tree, 'Árbol de Decisión')

# Crear DataFrame de comparación
comparison_metrics = pd.DataFrame([metrics_knn, metrics_tree])
comparison_metrics = comparison_metrics.round(4)

print("MÉTRICAS COMPARATIVAS:")
comparison_metrics
```

```python
print("\n🔍 ANÁLISIS DE MATRICES DE CONFUSIÓN")
print("="*60)

# Matrices de confusión
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_tree = confusion_matrix(y_test, y_pred_tree)

# Visualización mejorada de matrices de confusión
plt.figure(figsize=(15, 6))

# Matriz KNN
plt.subplot(1, 2, 1)
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'],
            cbar_kws={'label': 'Número de casos'})
plt.title(f'Matriz de Confusión - KNN\nExactitud: {metrics_knn["Exactitud"]:.3f}')
plt.ylabel('Valores Reales')
plt.xlabel('Predicciones')

# Añadir porcentajes
for i in range(2):
    for j in range(2):
        percentage = cm_knn[i, j] / cm_knn.sum() * 100
        plt.text(j+0.5, i+0.7, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=10, color='red')

# Matriz Árbol
plt.subplot(1, 2, 2)
sns.heatmap(cm_tree, annot=True, fmt='d', cmap='Greens',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'],
            cbar_kws={'label': 'Número de casos'})
plt.title(f'Matriz de Confusión - Árbol de Decisión\nExactitud: {metrics_tree["Exactitud"]:.3f}')
plt.ylabel('Valores Reales')
plt.xlabel('Predicciones')

# Añadir porcentajes
for i in range(2):
    for j in range(2):
        percentage = cm_tree[i, j] / cm_tree.sum() * 100
        plt.text(j+0.5, i+0.7, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=10, color='red')

plt.tight_layout()
plt.show()

# Análisis detallado de matrices
print("\nANÁLISIS DETALLADO DE MATRICES DE CONFUSIÓN:")
print("-" * 50)

def analizar_matriz_confusion(cm, modelo_nombre):
    tn, fp, fn, tp = cm.ravel()
    total = cm.sum()

    print(f"\n{modelo_nombre}:")
    print(f"  Verdaderos Negativos (TN): {tn} ({tn/total*100:.1f}%)")
    print(f"  Falsos Positivos (FP): {fp} ({fp/total*100:.1f}%)")
    print(f"  Falsos Negativos (FN): {fn} ({fn/total*100:.1f}%)")
    print(f"  Verdaderos Positivos (TP): {tp} ({tp/total*100:.1f}%)")

    # Errores más críticos
    if fp > fn:
        print(f"  ⚠️  Mayor problema: Falsos Positivos (predice churn cuando no hay)")
    elif fn > fp:
        print(f"  ⚠️  Mayor problema: Falsos Negativos (no detecta churn real)")
    else:
        print(f"  ✅ Errores balanceados")

analizar_matriz_confusion(cm_knn, "KNN")
analizar_matriz_confusion(cm_tree, "Árbol de Decisión")
```

```python
print("\n🔬 ANÁLISIS DE OVERFITTING/UNDERFITTING")
print("="*60)

# Validación cruzada para detectar overfitting
cv_scores_knn = cross_val_score(knn_model, X_train_scaled, y_train, cv=5, scoring='accuracy')
cv_scores_tree = cross_val_score(tree_model, X_train, y_train, cv=5, scoring='accuracy')

# Accuracy en entrenamiento vs test
train_accuracy_knn = knn_model.score(X_train_scaled, y_train)
test_accuracy_knn = accuracy_score(y_test, y_pred_knn)

train_accuracy_tree = tree_model.score(X_train, y_train)
test_accuracy_tree = accuracy_score(y_test, y_pred_tree)

# Crear DataFrame de análisis
overfitting_analysis = pd.DataFrame({
    'Modelo': ['KNN', 'Árbol de Decisión'],
    'Train Accuracy': [train_accuracy_knn, train_accuracy_tree],
    'Test Accuracy': [test_accuracy_knn, test_accuracy_tree],
    'CV Mean': [cv_scores_knn.mean(), cv_scores_tree.mean()],
    'CV Std': [cv_scores_knn.std(), cv_scores_tree.std()],
    'Gap (Train-Test)': [train_accuracy_knn - test_accuracy_knn,
                        train_accuracy_tree - test_accuracy_tree]
})

overfitting_analysis = overfitting_analysis.round(4)
print("ANÁLISIS DE OVERFITTING/UNDERFITTING:")
overfitting_analysis
```

```python
# Visualización del análisis
plt.figure(figsize=(15, 5))

# Gráfico 1: Train vs Test Accuracy
plt.subplot(1, 3, 1)
x_pos = range(len(overfitting_analysis))
plt.bar([p - 0.2 for p in x_pos], overfitting_analysis['Train Accuracy'],
        width=0.4, label='Train Accuracy', alpha=0.7, color='lightblue')
plt.bar([p + 0.2 for p in x_pos], overfitting_analysis['Test Accuracy'],
        width=0.4, label='Test Accuracy', alpha=0.7, color='lightcoral')
plt.xlabel('Modelos')
plt.ylabel('Accuracy')
plt.title('Train vs Test Accuracy')
plt.xticks(x_pos, overfitting_analysis['Modelo'], rotation=45)
plt.legend()
plt.ylim(0, 1)

# Gráfico 2: Gap Analysis
plt.subplot(1, 3, 2)
colors = ['red' if gap > 0.05 else 'yellow' if gap > 0.02 else 'green'
         for gap in overfitting_analysis['Gap (Train-Test)']]
plt.bar(x_pos, overfitting_analysis['Gap (Train-Test)'],
        color=colors, alpha=0.7)
plt.xlabel('Modelos')
plt.ylabel('Gap (Train - Test)')
plt.title('Gap de Accuracy\n(Indicador de Overfitting)')
plt.xticks(x_pos, overfitting_analysis['Modelo'], rotation=45)
plt.axhline(y=0.05, color='red', linestyle='--', alpha=0.5, label='Umbral crítico')
plt.legend()

# Gráfico 3: Cross Validation
plt.subplot(1, 3, 3)
plt.bar(x_pos, overfitting_analysis['CV Mean'],
        yerr=overfitting_analysis['CV Std'], capsize=5,
        color='lightgreen', alpha=0.7)
plt.xlabel('Modelos')
plt.ylabel('CV Accuracy (Mean ± Std)')
plt.title('Validación Cruzada\n(5-Fold)')
plt.xticks(x_pos, overfitting_analysis['Modelo'], rotation=45)

plt.tight_layout()
plt.show()
```

```python
print("\n🎯 ANÁLISIS CRÍTICO Y COMPARACIÓN")
print("="*60)

# Determinar el mejor modelo
best_f1 = comparison_metrics.loc[comparison_metrics['F1-Score'].idxmax()]
best_accuracy = comparison_metrics.loc[comparison_metrics['Exactitud'].idxmax()]
best_auc = comparison_metrics.loc[comparison_metrics['ROC-AUC'].idxmax()]

print("MEJOR MODELO POR MÉTRICA:")
print(f"📊 Mejor F1-Score: {best_f1['Modelo']} ({best_f1['F1-Score']:.4f})")
print(f"🎯 Mejor Exactitud: {best_accuracy['Modelo']} ({best_accuracy['Exactitud']:.4f})")
print(f"📈 Mejor ROC-AUC: {best_auc['Modelo']} ({best_auc['ROC-AUC']:.4f})")

# Análisis de overfitting
print(f"\nANÁLISIS DE OVERFITTING:")
for idx, row in overfitting_analysis.iterrows():
    modelo = row['Modelo']
    gap = row['Gap (Train-Test)']
    cv_std = row['CV Std']

    if gap > 0.05:
        status = "🔴 OVERFITTING DETECTADO"
        recommendation = "Reducir complejidad, regularización, más datos"
    elif gap > 0.02:
        status = "🟡 POSIBLE OVERFITTING"
        recommendation = "Monitorear, considerar regularización"
    else:
        status = "🟢 BUEN AJUSTE"
        recommendation = "Modelo estable"

    if cv_std > 0.05:
        stability = "Inestable (alta variabilidad)"
    else:
        stability = "Estable"

    print(f"\n{modelo}:")
    print(f"  Estado: {status}")
    print(f"  Gap Train-Test: {gap:.4f}")
    print(f"  Estabilidad: {stability} (CV Std: {cv_std:.4f})")
    print(f"  Recomendación: {recommendation}")

# =========================================================================
# RECOMENDACIONES FINALES
# =========================================================================

print(f"\n💡 RECOMENDACIONES FINALES")
print("="*60)

# Calcular score general (promedio de métricas normalizadas)
comparison_metrics['Score_General'] = (
    comparison_metrics['F1-Score'] * 0.3 +
    comparison_metrics['ROC-AUC'] * 0.3 +
    comparison_metrics['Precisión'] * 0.2 +
    comparison_metrics['Recall'] * 0.2
)

mejor_modelo_general = comparison_metrics.loc[comparison_metrics['Score_General'].idxmax(), 'Modelo']

print(f"🏆 MODELO RECOMENDADO: {mejor_modelo_general}")
print(f"📊 Score General: {comparison_metrics.loc[comparison_metrics['Score_General'].idxmax(), 'Score_General']:.4f}")

# Recomendaciones específicas por contexto de negocio
print(f"\nRECOMENDACIONES POR CONTEXTO:")

if comparison_metrics.loc[comparison_metrics['Modelo'] == 'KNN', 'Recall'].values[0] > \
   comparison_metrics.loc[comparison_metrics['Modelo'] == 'Árbol de Decisión', 'Recall'].values[0]:
    print("🎯 Para DETECTAR MÁS CHURN (minimizar falsos negativos): KNN")
else:
    print("🎯 Para DETECTAR MÁS CHURN (minimizar falsos negativos): Árbol de Decisión")

if comparison_metrics.loc[comparison_metrics['Modelo'] == 'KNN', 'Precisión'].values[0] > \
   comparison_metrics.loc[comparison_metrics['Modelo'] == 'Árbol de Decisión', 'Precisión'].values[0]:
    print("💰 Para EVITAR FALSOS POSITIVOS (precisión en campañas): KNN")
else:
    print("💰 Para EVITAR FALSOS POSITIVOS (precisión en campañas): Árbol de Decisión")

print("🔍 Para INTERPRETABILIDAD: Árbol de Decisión (muestra importancia de variables)")
print("⚡ Para VELOCIDAD DE PREDICCIÓN: Árbol de Decisión (no requiere normalización)")

print(f"\n🚀 PRÓXIMOS PASOS SUGERIDOS:")
print("1. Probar otros algoritmos (Random Forest, Gradient Boosting)")
print("2. Optimizar hiperparámetros con GridSearchCV")
print("3. Aplicar técnicas de balanceo de clases si es necesario")
print("4. Realizar feature engineering adicional")
print("5. Validar con datos más recientes")
```

# Informe Detallado: Factores de Churn y Estrategias de Retención

Este informe presenta un análisis de los datos de clientes de telecomunicaciones para identificar los principales factores que influyen en la cancelación (Churn) y propone estrategias de retención basadas en los hallazgos.

## 1. Resumen del Análisis

Se ha realizado un análisis exploratorio de datos, preprocesamiento y entrenamiento de dos modelos de clasificación (KNN y Árbol de Decisión) para predecir el Churn de clientes. El conjunto de datos presenta un desbalance significativo, con un 26.58% de clientes que cancelaron y un 73.42% que no.

## 2. Factores Clave que Influyen en el Churn

El análisis de correlación y el rendimiento de los modelos han revelado que los siguientes factores tienen una influencia significativa en la probabilidad de que un cliente cancele:

*   **Tipo de Contrato (Contract_Month-to-month):** Los clientes con contratos mes a mes tienen una correlación positiva fuerte con el Churn (0.405), lo que indica que son mucho más propensos a cancelar en comparación con aquellos con contratos de uno o dos años.
*   **Antigüedad (tenure):** Existe una correlación negativa fuerte con el Churn (-0.354), lo que sugiere que los clientes con mayor antigüedad son menos propensos a cancelar. Los clientes que cancelan suelen tener una antigüedad menor (como se observa en el boxplot de tenure vs Churn).
*   **Seguridad Online (OnlineSecurity_No):** La ausencia de seguridad online tiene una correlación positiva considerable con el Churn (0.342). Los clientes que no tienen este servicio son más propensos a irse.
*   **Soporte Técnico (TechSupport_No):** Similar a la seguridad online, no contar con soporte técnico muestra una correlación positiva importante con el Churn (0.337).
*   **Servicio de Internet (InternetService_Fiber optic y InternetService_No):** Tener servicio de fibra óptica tiene una correlación positiva con el Churn (0.307), mientras que no tener servicio de internet tiene una correlación negativa (-0.228). Esto podría indicar problemas con la calidad o el costo de la fibra óptica, o que los clientes sin internet tienen menos motivos para cancelar servicios de telecomunicaciones.
*   **Contrato de Dos Años (Contract_Two year):** Este tipo de contrato tiene una fuerte correlación negativa con el Churn (-0.302), reafirmando que los contratos a largo plazo fomentan la retención.
*   **Método de Pago (PaymentMethod_Electronic check):** El uso de cheques electrónicos tiene una correlación positiva con el Churn (0.301).

Otras variables como `OnlineBackup_No`, `DeviceProtection_No`, `PaperlessBilling_Yes`, `Charges.Monthly` y `Charges.Total` también mostraron cierta correlación, aunque en menor medida.

## 3. Análisis de Modelos de Clasificación

Se evaluaron dos modelos: K-Nearest Neighbors (KNN) y Árbol de Decisión.

*   **KNN:** Obtuvo una mejor exactitud general (0.7564) y precisión (0.5436). Sin embargo, su recall (0.5223) fue menor, lo que significa que identifica menos clientes que realmente cancelan (Falsos Negativos altos: 12.7%). Presenta overfitting (Gap Train-Test: 0.0786).
*   **Árbol de Decisión:** Aunque su exactitud fue menor (0.7261), mostró un recall significativamente más alto (0.7487) y un ROC-AUC superior (0.7922). Esto indica que es mejor detectando a los clientes que cancelarán (Falsos Negativos bajos: 6.7%), aunque a costa de una menor precisión (Falsos Positivos altos: 20.7%). También muestra signos de overfitting (Gap Train-Test: 0.0704).

**Conclusión de Modelos:** El Árbol de Decisión se considera el **modelo recomendado** por su mejor capacidad para identificar el Churn real (recall y ROC-AUC) y su interpretabilidad. Sin embargo, ambos modelos muestran overfitting, lo que requiere ajustes.

## 4. Estrategias de Retención Propuestas

Basadas en los factores de influencia identificados y el análisis de modelos, se proponen las siguientes estrategias:

*   **Promover Contratos a Largo Plazo:** Ofrecer incentivos significativos (descuentos, servicios adicionales) a los clientes con contratos mes a mes para que cambien a contratos de uno o dos años.
*   **Mejorar Servicios Adicionales:** Enfocarse en mejorar la adopción y satisfacción con los servicios de Seguridad Online y Soporte Técnico. Promocionar sus beneficios y resolver activamente los problemas de los clientes que no los tienen.
*   **Analizar y Mejorar la Experiencia con Fibra Óptica:** Investigar las causas de la alta correlación entre el servicio de fibra óptica y el Churn. Podría ser calidad del servicio, problemas de instalación, costos o atención al cliente. Implementar mejoras específicas.
*   **Optimizar Métodos de Pago:** Analizar por qué los clientes que pagan con cheque electrónico son más propensos a cancelar. Podría haber problemas con el proceso, recordatorios de pago, o ser un indicador de un segmento de clientes con mayor riesgo.
*   **Programas de Fidelización para Clientes Antiguos:** Reconocer y recompensar a los clientes con alta antigüedad para fortalecer su lealtad y reducir la probabilidad de que consideren cancelar.
*   **Campañas de Retención Dirigidas:** Utilizar el modelo de Árbol de Decisión (o un modelo optimizado) para identificar a los clientes con mayor riesgo de Churn (alta probabilidad predicha). Implementar campañas de retención personalizadas para estos segmentos.
*   **Monitorear Clientes Nuevos:** Prestar especial atención a los clientes con baja antigüedad, ya que son más propensos a cancelar. Ofrecer un onboarding efectivo y soporte proactivo.
