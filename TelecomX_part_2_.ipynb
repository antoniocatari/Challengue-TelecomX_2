```python

# DivisiÃ³n de datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split

# DivisiÃ³n 70/30 (puedes cambiar a 80/20 si prefieres)
X_train, X_test, y_train, y_test = train_test_split(
    x,  # Variables independientes (ya transformadas)
    y,  # Variable dependiente (ya codificada)
    test_size=0.3,  # 30% para prueba, 70% para entrenamiento
    random_state=42,  # Para reproducibilidad
    stratify=y  # Mantiene la proporciÃ³n de clases en ambos conjuntos
)

# Verificar las dimensiones
print("Dimensiones de los conjuntos:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")

# Verificar la proporciÃ³n de clases en cada conjunto
train_proportion = pd.Series(y_train).value_counts(normalize=True).sort_index()
test_proportion = pd.Series(y_test).value_counts(normalize=True).sort_index()

print("\nProporciÃ³n de clases en entrenamiento:")
print(f"No Churn (0): {train_proportion[0]:.3f}")
print(f"Churn (1): {train_proportion[1]:.3f}")

print("\nProporciÃ³n de clases en prueba:")
print(f"No Churn (0): {test_proportion[0]:.3f}")
print(f"Churn (1): {test_proportion[1]:.3f}")

print(f"\nâœ… DivisiÃ³n completada exitosamente")
print(f"ðŸ“Š Total de muestras: {len(x)}")
print(f"ðŸŽ¯ Entrenamiento: {len(X_train)} muestras ({len(X_train)/len(x)*100:.1f}%)")
print(f"ðŸ§ª Prueba: {len(X_test)} muestras ({len(X_test)/len(x)*100:.1f}%)")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================================
# MODELO 1: KNN (K-Nearest Neighbors) - REQUIERE NORMALIZACIÃ“N
# =========================================================================

print("ðŸ¤– MODELO 1: K-NEAREST NEIGHBORS (KNN)")
print("="*50)

# Normalizar los datos para KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo KNN
knn_model = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_model.fit(X_train_scaled, y_train)

# Predicciones
y_pred_knn = knn_model.predict(X_test_scaled)
y_pred_proba_knn = knn_model.predict_proba(X_test_scaled)[:, 1]

# MÃ©tricas KNN
accuracy_knn = accuracy_score(y_test, y_pred_knn)
roc_auc_knn = roc_auc_score(y_test, y_pred_proba_knn)

print(f"Exactitud (Accuracy): {accuracy_knn:.4f}")
print(f"ROC-AUC Score: {roc_auc_knn:.4f}")
print("\nReporte de clasificaciÃ³n KNN:")
print(classification_report(y_test, y_pred_knn))
```

```python
print("\nðŸŒ³ MODELO 2: ÃRBOL DE DECISIÃ“N")
print("="*50)

# Crear y entrenar el modelo de Ãrbol de DecisiÃ³n
tree_model = DecisionTreeClassifier(
    random_state=42,
    max_depth=10,  # Evita overfitting
    min_samples_split=20,  # MÃ­nimo de muestras para dividir
    min_samples_leaf=10,   # MÃ­nimo de muestras por hoja
    class_weight='balanced'  # Balancea las clases automÃ¡ticamente
)
tree_model.fit(X_train, y_train)  # Sin normalizaciÃ³n

# Predicciones
y_pred_tree = tree_model.predict(X_test)
y_pred_proba_tree = tree_model.predict_proba(X_test)[:, 1]

# MÃ©tricas Ãrbol de DecisiÃ³n
accuracy_tree = accuracy_score(y_test, y_pred_tree)
roc_auc_tree = roc_auc_score(y_test, y_pred_proba_tree)

print(f"Exactitud (Accuracy): {accuracy_tree:.4f}")
print(f"ROC-AUC Score: {roc_auc_tree:.4f}")
print("\nReporte de clasificaciÃ³n Ãrbol de DecisiÃ³n:")
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
def evaluar_modelo(y_true, y_pred, y_pred_proba, modelo_nombre):
    """
    Calcula todas las mÃ©tricas de evaluaciÃ³n para un modelo
    """
    metrics = {
        'Modelo': modelo_nombre,
        'Exactitud': accuracy_score(y_true, y_pred),
        'PrecisiÃ³n': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred),
        'ROC-AUC': roc_auc_score(y_true, y_pred_proba)
    }
    return metrics
```

```python
print("ðŸ“Š EVALUACIÃ“N DETALLADA DE MODELOS")
print("="*60)

# Calcular mÃ©tricas para KNN
metrics_knn = evaluar_modelo(y_test, y_pred_knn, y_pred_proba_knn, 'KNN')

# Calcular mÃ©tricas para Ãrbol de DecisiÃ³n
metrics_tree = evaluar_modelo(y_test, y_pred_tree, y_pred_proba_tree, 'Ãrbol de DecisiÃ³n')

# Crear DataFrame de comparaciÃ³n
comparison_metrics = pd.DataFrame([metrics_knn, metrics_tree])
comparison_metrics = comparison_metrics.round(4)

print("MÃ‰TRICAS COMPARATIVAS:")
comparison_metrics
```

```python
print("\nðŸ” ANÃLISIS DE MATRICES DE CONFUSIÃ“N")
print("="*60)

# Matrices de confusiÃ³n
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_tree = confusion_matrix(y_test, y_pred_tree)

# VisualizaciÃ³n mejorada de matrices de confusiÃ³n
plt.figure(figsize=(15, 6))

# Matriz KNN
plt.subplot(1, 2, 1)
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'],
            cbar_kws={'label': 'NÃºmero de casos'})
plt.title(f'Matriz de ConfusiÃ³n - KNN\nExactitud: {metrics_knn["Exactitud"]:.3f}')
plt.ylabel('Valores Reales')
plt.xlabel('Predicciones')

# AÃ±adir porcentajes
for i in range(2):
    for j in range(2):
        percentage = cm_knn[i, j] / cm_knn.sum() * 100
        plt.text(j+0.5, i+0.7, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=10, color='red')

# Matriz Ãrbol
plt.subplot(1, 2, 2)
sns.heatmap(cm_tree, annot=True, fmt='d', cmap='Greens',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'],
            cbar_kws={'label': 'NÃºmero de casos'})
plt.title(f'Matriz de ConfusiÃ³n - Ãrbol de DecisiÃ³n\nExactitud: {metrics_tree["Exactitud"]:.3f}')
plt.ylabel('Valores Reales')
plt.xlabel('Predicciones')

# AÃ±adir porcentajes
for i in range(2):
    for j in range(2):
        percentage = cm_tree[i, j] / cm_tree.sum() * 100
        plt.text(j+0.5, i+0.7, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=10, color='red')

plt.tight_layout()
plt.show()

# AnÃ¡lisis detallado de matrices
print("\nANÃLISIS DETALLADO DE MATRICES DE CONFUSIÃ“N:")
print("-" * 50)

def analizar_matriz_confusion(cm, modelo_nombre):
    tn, fp, fn, tp = cm.ravel()
    total = cm.sum()

    print(f"\n{modelo_nombre}:")
    print(f"  Verdaderos Negativos (TN): {tn} ({tn/total*100:.1f}%)")
    print(f"  Falsos Positivos (FP): {fp} ({fp/total*100:.1f}%)")
    print(f"  Falsos Negativos (FN): {fn} ({fn/total*100:.1f}%)")
    print(f"  Verdaderos Positivos (TP): {tp} ({tp/total*100:.1f}%)")

    # Errores mÃ¡s crÃ­ticos
    if fp > fn:
        print(f"  âš ï¸  Mayor problema: Falsos Positivos (predice churn cuando no hay)")
    elif fn > fp:
        print(f"  âš ï¸  Mayor problema: Falsos Negativos (no detecta churn real)")
    else:
        print(f"  âœ… Errores balanceados")

analizar_matriz_confusion(cm_knn, "KNN")
analizar_matriz_confusion(cm_tree, "Ãrbol de DecisiÃ³n")
```

```python
print("\nðŸ”¬ ANÃLISIS DE OVERFITTING/UNDERFITTING")
print("="*60)

# ValidaciÃ³n cruzada para detectar overfitting
cv_scores_knn = cross_val_score(knn_model, X_train_scaled, y_train, cv=5, scoring='accuracy')
cv_scores_tree = cross_val_score(tree_model, X_train, y_train, cv=5, scoring='accuracy')

# Accuracy en entrenamiento vs test
train_accuracy_knn = knn_model.score(X_train_scaled, y_train)
test_accuracy_knn = accuracy_score(y_test, y_pred_knn)

train_accuracy_tree = tree_model.score(X_train, y_train)
test_accuracy_tree = accuracy_score(y_test, y_pred_tree)

# Crear DataFrame de anÃ¡lisis
overfitting_analysis = pd.DataFrame({
    'Modelo': ['KNN', 'Ãrbol de DecisiÃ³n'],
    'Train Accuracy': [train_accuracy_knn, train_accuracy_tree],
    'Test Accuracy': [test_accuracy_knn, test_accuracy_tree],
    'CV Mean': [cv_scores_knn.mean(), cv_scores_tree.mean()],
    'CV Std': [cv_scores_knn.std(), cv_scores_tree.std()],
    'Gap (Train-Test)': [train_accuracy_knn - test_accuracy_knn,
                        train_accuracy_tree - test_accuracy_tree]
})

overfitting_analysis = overfitting_analysis.round(4)
print("ANÃLISIS DE OVERFITTING/UNDERFITTING:")
overfitting_analysis
```

```python
# VisualizaciÃ³n del anÃ¡lisis
plt.figure(figsize=(15, 5))

# GrÃ¡fico 1: Train vs Test Accuracy
plt.subplot(1, 3, 1)
x_pos = range(len(overfitting_analysis))
plt.bar([p - 0.2 for p in x_pos], overfitting_analysis['Train Accuracy'],
        width=0.4, label='Train Accuracy', alpha=0.7, color='lightblue')
plt.bar([p + 0.2 for p in x_pos], overfitting_analysis['Test Accuracy'],
        width=0.4, label='Test Accuracy', alpha=0.7, color='lightcoral')
plt.xlabel('Modelos')
plt.ylabel('Accuracy')
plt.title('Train vs Test Accuracy')
plt.xticks(x_pos, overfitting_analysis['Modelo'], rotation=45)
plt.legend()
plt.ylim(0, 1)

# GrÃ¡fico 2: Gap Analysis
plt.subplot(1, 3, 2)
colors = ['red' if gap > 0.05 else 'yellow' if gap > 0.02 else 'green'
         for gap in overfitting_analysis['Gap (Train-Test)']]
plt.bar(x_pos, overfitting_analysis['Gap (Train-Test)'],
        color=colors, alpha=0.7)
plt.xlabel('Modelos')
plt.ylabel('Gap (Train - Test)')
plt.title('Gap de Accuracy\n(Indicador de Overfitting)')
plt.xticks(x_pos, overfitting_analysis['Modelo'], rotation=45)
plt.axhline(y=0.05, color='red', linestyle='--', alpha=0.5, label='Umbral crÃ­tico')
plt.legend()

# GrÃ¡fico 3: Cross Validation
plt.subplot(1, 3, 3)
plt.bar(x_pos, overfitting_analysis['CV Mean'],
        yerr=overfitting_analysis['CV Std'], capsize=5,
        color='lightgreen', alpha=0.7)
plt.xlabel('Modelos')
plt.ylabel('CV Accuracy (Mean Â± Std)')
plt.title('ValidaciÃ³n Cruzada\n(5-Fold)')
plt.xticks(x_pos, overfitting_analysis['Modelo'], rotation=45)

plt.tight_layout()
plt.show()
```

```python
print("\nðŸŽ¯ ANÃLISIS CRÃTICO Y COMPARACIÃ“N")
print("="*60)

# Determinar el mejor modelo
best_f1 = comparison_metrics.loc[comparison_metrics['F1-Score'].idxmax()]
best_accuracy = comparison_metrics.loc[comparison_metrics['Exactitud'].idxmax()]
best_auc = comparison_metrics.loc[comparison_metrics['ROC-AUC'].idxmax()]

print("MEJOR MODELO POR MÃ‰TRICA:")
print(f"ðŸ“Š Mejor F1-Score: {best_f1['Modelo']} ({best_f1['F1-Score']:.4f})")
print(f"ðŸŽ¯ Mejor Exactitud: {best_accuracy['Modelo']} ({best_accuracy['Exactitud']:.4f})")
print(f"ðŸ“ˆ Mejor ROC-AUC: {best_auc['Modelo']} ({best_auc['ROC-AUC']:.4f})")

# AnÃ¡lisis de overfitting
print(f"\nANÃLISIS DE OVERFITTING:")
for idx, row in overfitting_analysis.iterrows():
    modelo = row['Modelo']
    gap = row['Gap (Train-Test)']
    cv_std = row['CV Std']

    if gap > 0.05:
        status = "ðŸ”´ OVERFITTING DETECTADO"
        recommendation = "Reducir complejidad, regularizaciÃ³n, mÃ¡s datos"
    elif gap > 0.02:
        status = "ðŸŸ¡ POSIBLE OVERFITTING"
        recommendation = "Monitorear, considerar regularizaciÃ³n"
    else:
        status = "ðŸŸ¢ BUEN AJUSTE"
        recommendation = "Modelo estable"

    if cv_std > 0.05:
        stability = "Inestable (alta variabilidad)"
    else:
        stability = "Estable"

    print(f"\n{modelo}:")
    print(f"  Estado: {status}")
    print(f"  Gap Train-Test: {gap:.4f}")
    print(f"  Estabilidad: {stability} (CV Std: {cv_std:.4f})")
    print(f"  RecomendaciÃ³n: {recommendation}")

# =========================================================================
# RECOMENDACIONES FINALES
# =========================================================================

print(f"\nðŸ’¡ RECOMENDACIONES FINALES")
print("="*60)

# Calcular score general (promedio de mÃ©tricas normalizadas)
comparison_metrics['Score_General'] = (
    comparison_metrics['F1-Score'] * 0.3 +
    comparison_metrics['ROC-AUC'] * 0.3 +
    comparison_metrics['PrecisiÃ³n'] * 0.2 +
    comparison_metrics['Recall'] * 0.2
)

mejor_modelo_general = comparison_metrics.loc[comparison_metrics['Score_General'].idxmax(), 'Modelo']

print(f"ðŸ† MODELO RECOMENDADO: {mejor_modelo_general}")
print(f"ðŸ“Š Score General: {comparison_metrics.loc[comparison_metrics['Score_General'].idxmax(), 'Score_General']:.4f}")

# Recomendaciones especÃ­ficas por contexto de negocio
print(f"\nRECOMENDACIONES POR CONTEXTO:")

if comparison_metrics.loc[comparison_metrics['Modelo'] == 'KNN', 'Recall'].values[0] > \
   comparison_metrics.loc[comparison_metrics['Modelo'] == 'Ãrbol de DecisiÃ³n', 'Recall'].values[0]:
    print("ðŸŽ¯ Para DETECTAR MÃS CHURN (minimizar falsos negativos): KNN")
else:
    print("ðŸŽ¯ Para DETECTAR MÃS CHURN (minimizar falsos negativos): Ãrbol de DecisiÃ³n")

if comparison_metrics.loc[comparison_metrics['Modelo'] == 'KNN', 'PrecisiÃ³n'].values[0] > \
   comparison_metrics.loc[comparison_metrics['Modelo'] == 'Ãrbol de DecisiÃ³n', 'PrecisiÃ³n'].values[0]:
    print("ðŸ’° Para EVITAR FALSOS POSITIVOS (precisiÃ³n en campaÃ±as): KNN")
else:
    print("ðŸ’° Para EVITAR FALSOS POSITIVOS (precisiÃ³n en campaÃ±as): Ãrbol de DecisiÃ³n")

print("ðŸ” Para INTERPRETABILIDAD: Ãrbol de DecisiÃ³n (muestra importancia de variables)")
print("âš¡ Para VELOCIDAD DE PREDICCIÃ“N: Ãrbol de DecisiÃ³n (no requiere normalizaciÃ³n)")

print(f"\nðŸš€ PRÃ“XIMOS PASOS SUGERIDOS:")
print("1. Probar otros algoritmos (Random Forest, Gradient Boosting)")
print("2. Optimizar hiperparÃ¡metros con GridSearchCV")
print("3. Aplicar tÃ©cnicas de balanceo de clases si es necesario")
print("4. Realizar feature engineering adicional")
print("5. Validar con datos mÃ¡s recientes")
```

# Informe Detallado: Factores de Churn y Estrategias de RetenciÃ³n

Este informe presenta un anÃ¡lisis de los datos de clientes de telecomunicaciones para identificar los principales factores que influyen en la cancelaciÃ³n (Churn) y propone estrategias de retenciÃ³n basadas en los hallazgos.

## 1. Resumen del AnÃ¡lisis

Se ha realizado un anÃ¡lisis exploratorio de datos, preprocesamiento y entrenamiento de dos modelos de clasificaciÃ³n (KNN y Ãrbol de DecisiÃ³n) para predecir el Churn de clientes. El conjunto de datos presenta un desbalance significativo, con un 26.58% de clientes que cancelaron y un 73.42% que no.

## 2. Factores Clave que Influyen en el Churn

El anÃ¡lisis de correlaciÃ³n y el rendimiento de los modelos han revelado que los siguientes factores tienen una influencia significativa en la probabilidad de que un cliente cancele:

*   **Tipo de Contrato (Contract_Month-to-month):** Los clientes con contratos mes a mes tienen una correlaciÃ³n positiva fuerte con el Churn (0.405), lo que indica que son mucho mÃ¡s propensos a cancelar en comparaciÃ³n con aquellos con contratos de uno o dos aÃ±os.
*   **AntigÃ¼edad (tenure):** Existe una correlaciÃ³n negativa fuerte con el Churn (-0.354), lo que sugiere que los clientes con mayor antigÃ¼edad son menos propensos a cancelar. Los clientes que cancelan suelen tener una antigÃ¼edad menor (como se observa en el boxplot de tenure vs Churn).
*   **Seguridad Online (OnlineSecurity_No):** La ausencia de seguridad online tiene una correlaciÃ³n positiva considerable con el Churn (0.342). Los clientes que no tienen este servicio son mÃ¡s propensos a irse.
*   **Soporte TÃ©cnico (TechSupport_No):** Similar a la seguridad online, no contar con soporte tÃ©cnico muestra una correlaciÃ³n positiva importante con el Churn (0.337).
*   **Servicio de Internet (InternetService_Fiber optic y InternetService_No):** Tener servicio de fibra Ã³ptica tiene una correlaciÃ³n positiva con el Churn (0.307), mientras que no tener servicio de internet tiene una correlaciÃ³n negativa (-0.228). Esto podrÃ­a indicar problemas con la calidad o el costo de la fibra Ã³ptica, o que los clientes sin internet tienen menos motivos para cancelar servicios de telecomunicaciones.
